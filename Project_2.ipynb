```
from pyspark.sql.types import *

import pyspark.sql.functions as F

from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("my project 2")\
    .config("spark.kryoserializer.buffer.max", "512m")\
    .config('spark.jars.packages', 'org.apache.spark:spark-sql-kafka-0-10_2.12:3.2.0')\
        .getOrCreate()
sc = spark.sparkContext


for_students_demo_path = "/mnt/ddscoursedatastorage/ro/fwm-stb-data/proj_B_demographic"
demographic_df = spark.read.parquet(for_students_demo_path)

#demographic_df.cache()
display(demographic_df)
```

We would like to adjust and normalize the features:
1. For numerical variables - normalize each column according to the maximum value in that column
- so the values in each numerical column would be in the range of [0,1]
2. For categorical variables (non-ordinal) - use one hot encoding to change the values of the column
to a binary vector representation.
Then, combine all adjusted and normalized variables into one feature vector column that will represent
each household.

```
from pyspark.sql.functions import max, col
from pyspark.sql.types import FloatType
#normalaizing the numerical data to be [0-1]
demographic_df = demographic_df.withColumn("education_highest", demographic_df["education_highest"].cast(FloatType()))

max_house_size = demographic_df.select(col("household_size")).agg(max(col("household_size"))).first()[0]
max_num_adults = demographic_df.select(col("num_adults")).agg(max(col("num_adults"))).first()[0]
max_num_generations = demographic_df.select(col("num_generations")).agg(max(col("num_generations"))).first()[0]
max_length_resid = demographic_df.select(col("length_residence")).agg(max(col("length_residence"))).first()[0]
max_home_market_high = demographic_df.select(col("home_market_value")).agg(max(col("home_market_value"))).first()[0]
max_net_worth = demographic_df.select(col("net_worth")).agg(max(col("net_worth"))).first()[0]
max_education_high = demographic_df.select(col("education_highest")).agg(max(col("education_highest"))).first()[0]

demographic_df_normal_1 = demographic_df.withColumn("household_size", col("household_size")/max_house_size)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("num_adults", col("num_adults")/max_num_adults)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("num_generations", col("num_generations")/max_num_generations)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("length_residence", col("length_residence")/max_length_resid)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("home_market_value", col("home_market_value")/max_home_market_high)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("net_worth", col("net_worth")/max_net_worth)
demographic_df_normal_1 = demographic_df_normal_1.withColumn("education_highest", col("education_highest")/max_education_high)
```

```
from pyspark.ml.feature import OneHotEncoder
from pyspark.ml.feature import StringIndexer




# Define the columns to be one-hot encoded
columns_to_encode = ["marital_status", "race_code","dwelling_type","home_owner_status", "gender_individual"]

# Apply StringIndexer and OneHotEncoder for each column
indexers = [StringIndexer(inputCol=column, outputCol=column + "_idx")
            for column in columns_to_encode]



encoder = OneHotEncoder(inputCols=[column + "_idx" for column in columns_to_encode],
                        outputCols=[column + "_vec" for column in columns_to_encode])

# Fit and transform the DataFrame with indexers and encoder
indexer_models = [indexer.fit(demographic_df_normal_1) for indexer in indexers]
indexed_df = demographic_df_normal_1
for model in indexer_models:
    indexed_df = model.transform(indexed_df)

encoded_df = encoder.fit(indexed_df).transform(indexed_df)

encoded_df=encoded_df.select("household_size","num_adults","num_generations","length_residence","home_market_value","net_worth","education_highest","household_id","marital_status_vec","race_code_vec","dwelling_type_vec","home_owner_status_vec","gender_individual_vec")


```

```
from pyspark.ml.feature import VectorAssembler
from pyspark.sql.functions import col, expr


assembler = VectorAssembler(
    inputCols=["household_size","num_adults","num_generations","length_residence","home_market_value","net_worth","education_highest","race_code_vec","marital_status_vec","dwelling_type_vec","home_owner_status_vec","gender_individual_vec"],
    outputCol="features")

output2 = assembler.transform(encoded_df)
# output2.select("household_id","features").show()
# output2.show(truncate=False)


```

Now that we have one column that contains all our useful information, we can easily use many known
and good algorithms to infer information from our data.
We want to see if the data is naturally divided into several groups of households. In order to do that we
can use the PCA algorithm. Use the PCA algorithm to project the feature vectors from the previous part
on to 2 dimensional space and then plot the result in a scatter plot

```
import matplotlib.pyplot as plt
from pyspark.ml.feature import PCA
 
# Assuming you have the "output2" DataFrame available
 
# Create a PCA instance with k=2
pca = PCA(k=2, inputCol="features", outputCol="pcaFeatures")
 
# Fit the PCA model on the feature vectors
pcaModel = pca.fit(output2)
 
# Transform the feature vectors using the PCA model
pcaResult = pcaModel.transform(output2).select("pcaFeatures")
 
# Collect the results as a list of Row objects
rows = pcaResult.collect()

# Extract the two PCA columns for scatter plot
x = [row.pcaFeatures[0] for row in rows]
y = [row.pcaFeatures[1] for row in rows]

# Plot the scatter plot using Matplotlib
plt.scatter(x=x, y=y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of PCA')
plt.show()
```

```

from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors,VectorUDT
from pyspark.sql.functions import udf
from pyspark.sql.types import DoubleType
from pyspark.sql.window import Window
input2=output2

# Prepare the data for clustering
assembler = VectorAssembler(inputCols=["features"], outputCol="feature_vector")
assembledDF = assembler.transform(input2)




# Train a k-means model with k = 2
k = 6
kmeans = KMeans().setK(k).setFeaturesCol("feature_vector").setPredictionCol("cluster").setSeed(3)

model = kmeans.fit(assembledDF)

# Get the cluster assignments and distances for each household_id
predictions = model.transform(assembledDF)
centroids = model.clusterCenters()

# # Function to calculate distance from a vector to its assigned centroid
def getCentroid(cluster):
    return Vectors.dense(centroids[cluster])

getCentroidUDF = udf(getCentroid, VectorUDT())

# Select relevant columns and add the centroid column
result = predictions.select("household_id", "cluster","feature_vector").withColumn("centroid", getCentroidUDF("cluster"))


# Define a UDF to calculate the distance between a feature vector and its centroid
calculateDistanceUDF = udf(lambda features, centroid: float(features.squared_distance(centroid)), returnType=DoubleType())


# Add a new column with the distance from each household ID to its centroid
result = result.withColumn("distance", calculateDistanceUDF("feature_vector", "centroid"))

# Show the result
result.show()



```

```

from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

#define window over clusters distance
windowSpec  = Window.partitionBy("cluster").orderBy("distance")

# Get the full data for each cluster
full_data=result.withColumn("row_number",row_number().over(windowSpec))
  
# Filter the rows to get the 3rds subset
subset_3rds = full_data.filter(col("row_number") % 3 == 0)

# Filter the rows to get the 17ths subset
subset_17ths = full_data.filter(col("row_number") % 17 == 0)




```

```

kafka_server = "kafka96224.eastus.cloudapp.azure.com:29092" 
topic='viewstatic'
OFFSETS_PER_TRIGGER = 50000
SCHEMA = "device_id STRING, event_date STRING, event_time STRING, station_num STRING, prog_code STRING, household_id LONG"

static_df = spark.read\
                  .format("kafka")\
                  .option("kafka.bootstrap.servers", kafka_server)\
                  .option("subscribe", topic)\
                  .option("startingOffsets", "earliest")\
                  .option("failOnDataLoss",False)\
                  .load()

static_view_data = static_df.select(F.from_csv(F.decode("value", "US-ASCII"), schema=SCHEMA).alias("value")).select("value.*").na.drop()


static_view_data.cache()
```

```
from pyspark.sql.functions import count
clusters = []
static_view_data1=static_view_data.select("station_num","household_id")
static_view_data_grouped=static_view_data1.groupBy("station_num","household_id").agg(count("*").alias("viewing_count"))
```

```
from pyspark.sql.functions import count,sum
clusters_part2=[]
for i in range(6):
    #filter the clusters to subsets of row counts
    cluster_df = full_data.filter(col("cluster") == i)
    cluster_df=cluster_df.join(static_view_data_grouped,on="household_id",how="inner")
    _3rd_cluster=cluster_df.filter(col("row_number")%3 ==0)
    _17ths_cluster=cluster_df.filter(col("row_number")%17 ==0)
    #full
    cluster_df=cluster_df.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
    total_sum=cluster_df.agg(sum(col("sum_count"))).collect()[0][0]
    cluster_df_part2=cluster_df.withColumn("precentage",(100*col("sum_count")/total_sum))                        
    clusters.append(cluster_df)
    clusters_part2.append(cluster_df_part2)

    #the 3rth subset
    cluster_df=_3rd_cluster.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
    total_sum=cluster_df.agg(sum(col("sum_count"))).collect()[0][0]
    cluster_df_part2=cluster_df.withColumn("precentage",(100*col("sum_count")/total_sum))                        
    clusters.append(cluster_df)
    clusters_part2.append(cluster_df_part2)

    #the 17ths
    cluster_df=_17ths_cluster.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
    total_sum=cluster_df.agg(sum(col("sum_count"))).collect()[0][0]
    cluster_df_part2=cluster_df.withColumn("precentage",(100*col("sum_count")/total_sum))                     
    clusters.append(cluster_df)
    clusters_part2.append(cluster_df_part2)
    
    print(i)









```

```
#3. Now, repeat steps 1 and 2 - but for the entire data, not
# 
# on a per cluster basis. This is the popularity
# rating of each station among the general population.
# The output of this stage is the popularity ratings (percentages) of each station among the general
# public.

from pyspark.sql.functions import col, sum

part_3 = static_view_data_grouped
cluster_df=part_3.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
total_sum=cluster_df.agg(sum(col("sum_count"))).first()[0]
cluster_df_part3=cluster_df.withColumn("precentage_overall",(100*col("sum_count")/total_sum))                        

```

```
from pyspark.sql.functions import col, count, sum, expr
from pyspark.sql import DataFrame

big_df = None
batch_counter = 0
query = None
TO_RUN = True


def handle_batch(data, epoch_num):
    global big_df, batch_counter, streaming_df, TO_RUN,query
    
    if query is None:
        #Start the streaming query and assign it to the 'query' variable
        query = streaming_df.writeStream.format("console").start()

    if TO_RUN:
        clusters = []
        static_view_data1 = data.select("station_num", "household_id")
        static_view_data_grouped = static_view_data1.groupBy("station_num", "household_id").agg(
            count("*").alias("viewing_count"))
        
        if batch_counter == 0:
            big_df = static_view_data_grouped
            part3 = static_view_data_grouped
            cluster = part3.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
            total_sum = cluster.agg(sum(col("sum_count"))).first()[0]
            part3 = cluster.withColumn("precentage_overall", (100 * col("sum_count") / total_sum))
        else:
            big_df = big_df.union(static_view_data_grouped)
         

            cluster = big_df.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
            total_sum = cluster.agg(sum(col("sum_count"))).first()[0]
            part3 = cluster.withColumn("precentage_overall", (100 * col("sum_count") / total_sum))

        static_view_data_grouped = big_df
        clusters_partB = []

        print("batch number " + str(epoch_num + 1))
        # the code from previous

        for i in range(6):
            cluster_df = subset_3rds.filter(col("cluster") == i)
            cluster_df = cluster_df.join(static_view_data_grouped, on="household_id", how="inner")

            cluster_df = cluster_df.groupBy("station_num").agg(sum(col("viewing_count")).alias("sum_count"))
            total_sum = cluster_df.agg(sum(col("sum_count"))).first()[0]

            cluster_df_part2 = cluster_df.withColumn("percentage", (100 * col("sum_count") / total_sum))


            cluster_df_part2 = cluster_df_part2.join(part3, on="station_num", how="inner")
            cluster_df_part2 = cluster_df_part2.withColumn("diff_rank", expr("percentage - precentage_overall")).orderBy(
                expr("diff_rank").desc())

            clusters_partB.append(cluster_df_part2)
            print("cluster number " + str(i + 1))
            cluster_df_part2.select("station_num", "diff_rank").limit(7).show()

        # Increment the batch counter
        batch_counter += 1

        # Check if the desired number of batches is processed
        if batch_counter >= 5:
            TO_RUN = False
            query.stop()  # Stop the streaming query
            raise Exception("Desired number of batches processed")  # Raise an exception to exit the program
            
    else:
        query.stop()  # Stop the streaming query
        raise Exception("TO_RUN is False")  # Raise an exception to exit the program


```

```
streaming_df.writeStream\
    .foreachBatch(handle_batch)\
    .start()\
    .awaitTermination()
```
